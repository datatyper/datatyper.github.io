```python
def isprime(x):
    for i in range(2,x):
        if x % i == 0:
            return False
    return True
```

```python
for i in range(20):
    print(i, isprime(i), sep='\t')
```

```python
with plt.xkcd():
    plt.figure(dpi=100)
    plt.plot(x, [i**2 for i in x], 'ro')
    plt.plot(x, [i**3/5 for i in x], 'b--')
    plt.xlabel('The x axis')
    plt.ylabel('The y axis')
    plt.legend(('$x^2$', '$x^3/5$'))
    plt.title('Comparing $x^2$ with $x^3$')
    plt.show()
```
